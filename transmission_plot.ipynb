```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

Oe = 1000/(4*np.pi)


# Read the CSV file into a DataFrame
df = pd.read_csv('Transmissions.csv')
df["Transverse field"]=df["Transverse field"]/Oe
```

# T normalized

```python
# Group the dataframe by "Transverse field"
groups = df.groupby("Frequency")

# Iterate over each group and plot the scatter plot
for name, group in groups:
    plt.plot(group["Transverse field"], group["T1z"]/group["T1z"].iloc[0], label=f"{round(name*1e-9,1)}Ghz")
plt.gca().set_prop_cycle(None)
for name, group in groups:
    plt.plot(group["Transverse field"], group["T2z"]/group["T2z"].iloc[0], label=f"{round(name*1e-9,1)}Ghz",linestyle='--')

# Set the x-axis label
plt.xlabel('H [Oe]')

# Set the y-axis label
plt.ylabel('$T_z(normalized)$')

# Add a legend
plt.legend()


plt.savefig('T1z(H).pdf')
# Show the plot
plt.show()

# Group the dataframe by "Transverse field"
groups = df.groupby("Frequency")

# Iterate over each group and plot the scatter plot
for name, group in groups:
    plt.plot(group["Transverse field"], group["T1y"]/group["T1y"].iloc[0], label=f"{round(name*1e-9,1)}Ghz")
plt.gca().set_prop_cycle(None)
for name, group in groups:
    plt.plot(group["Transverse field"], group["T2y"]/group["T2y"].iloc[0], label=f"{round(name*1e-9,1)}Ghz",linestyle='--')


# Set the x-axis label
plt.xlabel('H [Oe]')

# Set the y-axis label
plt.ylabel('$T_y(normalized)$')

# Add a legend
plt.legend()

plt.savefig('T1y(H).pdf')

# Show the plot
plt.show()

```
